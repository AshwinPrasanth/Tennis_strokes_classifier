```python
import cv2, os

base_path = "../data/forehand"
files = os.listdir(base_path)
print("Number of forehand videos:", len(files))

cap = cv2.VideoCapture(os.path.join(base_path, files[0]))
fps = cap.get(cv2.CAP_PROP_FPS)
frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

print(f"Resolution: {width}x{height}, FPS: {fps}, Frames: {frames}")
cap.release()
```
